```python
import csv, os, glob
import numpy as np
import tensorflow as tf
import features as features_lib
import tensorflow as tf
import yamnet_short

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

parent_dir = "D:\\datasets\\\cer_dataset_16k_flattened_resampled\\"
classes = os.listdir(parent_dir)
NUM_CLASSES=len(classes)
Yamnet_Short = yamnet_short.Yamnet_short(NUM_CLASSES)
yamnet_model = Yamnet_Short.model()
yamnet_model.load_weights("D:/Github/YAMNET/yamnet_short/training/20220408-171632/fold2/checkpoints/cp.ckpt")

import params as yamnet_params
def get_embedding(waveform):
  params = yamnet_params.Params(sample_rate=16000, patch_hop_seconds=0.5)
  waveform_padded = features_lib.pad_waveform(waveform, params)
  log_mel_spectrogram, features = features_lib.waveform_to_log_mel_spectrogram_patches(
      waveform_padded, params)
  return tf.expand_dims(features, axis=-1)
# return log_mel_spectrogram
```

```python
import matplotlib
matplotlib.use('TkAgg')

from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk #NavigationToolbar2TkAgg
from matplotlib.figure import Figure
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random
import numpy as np
import sys
import tkinter as tk
import time

# Setup channel info
FORMAT = pyaudio.paInt16 # data type formate
CHANNELS = 1 # Adjust to your number of channels
RATE = 16000 # Sample Rate
CHUNK = 1024 # Block Size
RECORD_SECONDS = 5 # Record time
WAVE_OUTPUT_FILENAME = "file.wav"

# Startup pyaudio instance
audio = pyaudio.PyAudio()
# %matplotlib qt

# start Recording
stream = audio.open(format=FORMAT, channels=CHANNELS,
                rate=RATE, input=True,
                frames_per_buffer=CHUNK)
print("recording...")

fig, axes = plt.subplots(nrows=1)
fig.show()
fig.canvas.draw()
root = tk.Tk()
canvas = FigureCanvasTkAgg(fig, master=root)

while 1==1:
    frames = []
    # Record for RECORD_SECONDS
    for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
        data = stream.read(CHUNK)
        frames.append(np.frombuffer(data, dtype=np.int16))
    numpydata = np.hstack(frames)
        # print(float_data)
    axes.plot(numpydata)    
    canvas.draw()
    canvas._tkcanvas.pack(side=tk.TOP, fill=tk.BOTH, expand=1)

# # Stop Recording
stream.stop_stream()
stream.close()
audio.terminate()
```

```python
import matplotlib
matplotlib.use('TkAgg')

from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk #NavigationToolbar2TkAgg
from matplotlib.figure import Figure
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random
import numpy as np
import sys
import tkinter as tk
# import Tkinter as tk
import time

def get_frame():
    frames = []
    # Record for RECORD_SECONDS
    for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
        data = stream.read(CHUNK)
        frames.append(np.frombuffer(data, dtype=np.int16))
    data = np.hstack(frames)
    float_data = (data - np.min(data)) / (np.max(data) - np.min(data)) *2-1
    return float_data
    
def fast_plot(fig, ax, data, classes):
    ax.cla()
    # ax.plot(data)
    # for (i, bar_data) in enumerate(data)
    ind = range(len(data))
    ax.bar(ind, data)
    ax.set_xticks(ind, labels=classes)
    ax.set_xticklabels(ax.get_xticks(), rotation = 90)
    return fig

def fast_imshow(fig, ax, melspec):
    ax.cla()
    ax.imshow(np.squeeze(melspec))
    return fig

def fast_wav_plot(fig, ax, wav):
    ax.cla()
    # ax.imshow(np.squeeze(melspec))
    ax.plot(wav)
    return fig

class App():
    def __init__(self):
        self.root = tk.Tk()
        self.root.wm_title("Embedding in TK")

        self.fig = plt.figure()
        self.ax1 = self.fig.add_subplot(211)
        self.ax1.set_xlabel('Predictions')
        self.ax1.set_ylabel('Classes')
        
        self.ax2 = self.fig.add_subplot(212)
        self.ax2.set_xlabel('Band')
        self.ax2.set_ylabel('Time')

        # self.fig = fast_plot(self.fig, self.ax, [0])

        self.canvas = FigureCanvasTkAgg(self.fig, master=self.root)
        self.toolbar = NavigationToolbar2Tk( self.canvas, self.root )
        self.toolbar.update()
        self.canvas._tkcanvas.pack(side=tk.TOP, fill=tk.BOTH, expand=1)

        self.label = tk.Label(text="")
        self.label.pack()
        self.update_clock()
        self.root.mainloop()

    def update_clock(self):
        data = get_frame()
        predictions=np.arange(7)
        # melspec = get_embedding(data)
        # predictions = yamnet_model.predict(melspec)[0]
        self.fig = fast_plot(self.fig,self.ax1, predictions, classes)
        # self.fig = fast_imshow(self.fig,self.ax2, melspec)
        self.fig = fast_wav_plot(self.fig,self.ax2, data)
        self.canvas.draw()
        self.canvas._tkcanvas.pack(side=tk.TOP, fill=tk.BOTH, expand=1)
        now = time.strftime("%H:%M:%S")
        # self.label.configure(text=now)
        self.label.configure(text=predictions)
        self.root.after(1000, self.update_clock)
```

```python
# Setup channel info
FORMAT = pyaudio.paInt16 # data type formate
CHANNELS = 1 # Adjust to your number of channels
RATE = 16000 # Sample Rate
CHUNK = 1024 # Block Size
RECORD_SECONDS = 1 # Record time
WAVE_OUTPUT_FILENAME = "file.wav"

# Startup pyaudio instance
audio = pyaudio.PyAudio()
# start Recording
stream = audio.open(format=FORMAT, channels=CHANNELS,
                rate=RATE, input=True,
                frames_per_buffer=CHUNK)
print("recording...")

app=App()

# # Stop Recording
stream.stop_stream()
stream.close()
audio.terminate()
```
